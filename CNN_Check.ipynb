```
image = plt.imread("5.jpg")
dst = cv2.resize(image, dsize=(28, 28), interpolation=cv2.INTER_AREA)
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
plt.imshow(~dst)
plt.show()

test_num = ~dst
test_num = (test_num > 150) * test_num
test_num = test_num.astype('float32') / 255.

plt.imshow(test_num, cmap='Greys', interpolation='nearest');

test_num = test_num.reshape((1, 28, 28, 1))

print(model.predict_classes(test_num))
print(model.predict(test_num))
```
